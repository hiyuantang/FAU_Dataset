```python
def transf(img):
    img_around = F.pad(img, (0,0,1,1,1,1), 'constant', 0)
    img_pos0 = F.pad(img, (0,0,0,2,2,0), 'constant', 0)
    #img_pos1 = F.pad(img, (0,0,0,2,0,2), 'constant', 0)
    img_final = img_around - img_pos0
    img_final = (img_final - img_final.min())/(img_final.max()-img_final.min())
    img_final = np.arctan(img_final - 0.5)
    img_final = np.abs(img_final+0.1)
    img_final = (img_final - img_final.min())/(img_final.max()-img_final.min())
    return img_final


img_0 = mpimg.imread('D:/Datasets/face_gen/images/African/a/4a_4.2.png')
img_1 = mpimg.imread('D:/Datasets/face_gen/images/African/aw/4aw_4.2.png')
img_0 = torch.tensor(img_0[100:700, 50:650, 0:3])
img_1 = torch.tensor(img_1[100:700, 50:650, 0:3])
img_0 = transf(img_0)
img_1 = transf(img_1)
print(((img_0 - img_1)**2).mean())
```
