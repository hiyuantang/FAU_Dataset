```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
import torch.nn as nn
import cv2
```

```python
def rgb(path): 
    image_path = path
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #print('image shape_0: '+str(image.shape), 'min: '+str(image.min()), 'max: '+str(image.max()), sep=' | ')
    image = torch.tensor(image[100:700, 50:650, 0:3])
    #print('image shape_1: '+str(image.shape), 'min: '+str(image.min()), 'max: '+str(image.max()), sep=' | ')
    image = (image - image.min())/(image.max()-image.min())
    #print('image shape_2: '+str(image.shape), 'min: '+str(image.min()), 'max: '+str(image.max()), sep=' | ')
    ###########
    r_channel = image[:, :, 0]
    r_channel[r_channel == 1.] = 0.
    g_channel = image[:, :, 1]
    g_channel[g_channel == 1.] = 0.
    b_channel = image[:, :, 2]
    b_channel[b_channel == 1.] = 0.
    print('image shape_3: '+str(image.shape), 'min: '+str(image.min()), 'max: '+str(image.max()), sep=' | ')
    ###########
    plt.imshow(image)
    print('image shape_3: '+str(image.shape), 
        'R_sum: '+str(r_channel.sum()), 
        'G_sum: '+str(g_channel.sum()), 
        'B_sum: '+str(b_channel.sum())+'\n', 
        'R/G: '+str(r_channel.sum()/g_channel.sum()),
        'G/B: '+str(g_channel.sum()/b_channel.sum()), sep=' | ')
```
